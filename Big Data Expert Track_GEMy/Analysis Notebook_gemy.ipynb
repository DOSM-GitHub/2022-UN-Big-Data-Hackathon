```python
fig = go.Figure()

overseas_cnt_df = comt
source_to_dest = zip(overseas_cnt_df["LatOrig"], overseas_cnt_df["Latitude"],
                     overseas_cnt_df["LongOrig"], overseas_cnt_df["Longitude"],
                     overseas_cnt_df["count"])

## Loop thorugh each flight entry
for slat,dlat, slon, dlon, num_flights in source_to_dest:
    fig.add_trace(go.Scattergeo(
                        lat = [slat,dlat],
                        lon = [slon, dlon],
                        mode = 'lines',
                        line = dict(width = num_flights/50, color="red")
                        ))

fig.update_layout(title_text = 'Connection Map Depicting Flights from Brazil to All Other Countries',
                  #height=700, width=900,
                  margin={"t":0,"b":0,"l":0, "r":0, "pad":0},
                  showlegend=False)

fig.show()
```
